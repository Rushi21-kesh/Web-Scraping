![](https://upload.wikimedia.org/wikipedia/commons/thumb/5/5c/Olympic_rings_without_rims.svg/1280px-Olympic_rings_without_rims.svg.png)

- The medal table for all [Olympic Games](https://en.wikipedia.org/wiki/All-time_Olympic_Games_medal_table) from 1896 to 2018, including Summer Olympic Games, Winter Olympic Games, and a combined total of both, is tabulated below. These Olympic medal counts do not include the 1906 Intercalated Games which are no longer recognized by the International Olympic Committee (IOC) as official Games.

- Medal totals in this table are current as of the [2018 Winter Olympics](https://en.wikipedia.org/wiki/2018_Winter_Olympics) in Pyeongchang County, South Korea, and [all changes in medal standings](https://en.wikipedia.org/wiki/List_of_stripped_Olympic_medals) due to doping cases and medal redistributions up to 25 November 2020 are taken into account

### Import required libraries

```python
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd
import requests
```

### Requestiong URL

```python
url = 'https://en.wikipedia.org/wiki/All-time_Olympic_Games_medal_table'
```

```python
source = urllib.request.urlopen(url)
```

### Creating object of BeautifulSoup

```python
soup = BeautifulSoup(source,'lxml')
```

```python
table=soup('table',class_="wikitable sortable")[0]
```

```python
data= []
for i in range(2,(len(table.find_all('tr'))-1)):
    for j in range(16):
        data.append(table.find_all('tbody')[0].find_all('tr')[i].find_all('td')[j].get_text())
```

```python
data = [data[i:i+16] for i in range(0,len(data),16)]
```

```python
dataset=pd.DataFrame(data,columns=['Team(IOC Code)','summer_rank','s_gold','s_silver','s_bronze','summer_total','winter_rank','w_gold','w_silver','w_bronze','winter_total','overall_rank','total_gold','total_silver','total_bronze','total_medals'])
```

```python
pd.set_option('display.max_rows', 500)
dataset
```

### Cleaning Dataset

```python
for col in dataset.columns:
    dataset[col]=dataset[col].str.replace('\n','')
```

### Saving a dataser to CSV format

```python
dataset.to_csv("Olympic Games medal Dataset(from 1896 to 2018).csv",index=False)
```
